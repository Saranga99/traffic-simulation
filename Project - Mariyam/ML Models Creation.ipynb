```python
import pandas as pd 
import numpy as np 
import sklearn 
from sklearn.model_selection import train_test_split
```

```python
High_dataset = pd.read_csv('High traffic.csv')
High_dataset.isnull().sum()
```

```python
Low_dataset = pd.read_csv('Low traffic.csv')
Low_dataset.isnull().sum()
```

```python
Medium_dataset = pd.read_csv('Medium traffic.csv')
Medium_dataset.isnull().sum()
```

```python
High_dataset.head()
```

```python
Low_dataset.head()
```

```python
Medium_dataset.head()
```

```python
High_dataset.insert(6, "Traffic","High", True)
High_dataset.head()
```

```python
Low_dataset.insert(6, "Traffic","Low", True)
Low_dataset.head()
```

```python
Medium_dataset.insert(6, "Traffic","Medium", True)
Medium_dataset.head()
```

```python
frames = [High_dataset, Medium_dataset, Low_dataset]

Traffic_dataset = pd.concat(frames)
Traffic_dataset = Traffic_dataset.drop(columns=["lorry","truck"])
Traffic_dataset
```

```python
#Encode the class variable
Traffic_Encode = {'Low':0, 'Medium':1,'High':2}
Traffic_dataset.Traffic = Traffic_dataset.Traffic.map(Traffic_Encode)
Traffic_dataset
```

```python
#features
X = Traffic_dataset.drop(['Traffic'], axis=1)

#Label
y = Traffic_dataset['Traffic']

#split the data set as training set and test set randomly
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7,test_size=0.3, random_state = 42)
```

```python
# import Random Forest classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_curve, roc_auc_score

# instantiate the classifier 
rfc = RandomForestClassifier(random_state=42)

# fit the model
rfc.fit(X_train, y_train)

# Predict the Test set results
y_pred = rfc.predict(X_test)

print("The accuracy score of Random Forest Classifier is {}".format(accuracy_score(y_test, y_pred)))
print()
print("Classification Report:")
print(classification_report(y_test, y_pred))
print()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_curve, roc_auc_score

gini = DecisionTreeClassifier(criterion = "entropy",
            random_state = 880,max_depth=3, min_samples_leaf=5)
            
gini.fit(X_train, y_train)

y_pred = gini.predict(X_test)

print("The accuracy score of DecisionTree Classifier is {}".format(accuracy_score(y_test, y_pred)))
print()
print("Classification Report:")
print(classification_report(y_test, y_pred))
print()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_curve, roc_auc_score

# instantiate the classifier 
gb = GradientBoostingClassifier()

# fit the model
gb.fit(X_train, y_train)

# Predict the Test set results
y_pred = gb.predict(X_test)

print("The accuracy score of GradientBoosting Classifier is {}".format(accuracy_score(y_test, y_pred)))
print()
print("Classification Report:")
print(classification_report(y_test, y_pred))
print()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_curve, roc_auc_score

# instantiate the classifier 
knn = KNeighborsClassifier(n_neighbors=5)

# fit the model
knn.fit(X_train, y_train)

# Predict the Test set results
y_pred = knn.predict(X_test)

print("The accuracy score of KNeighbors Classifier is {}".format(accuracy_score(y_test, y_pred)))
print()
print("Classification Report:")
print(classification_report(y_test, y_pred))
print()
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score, classification_report, roc_curve, roc_auc_score

# instantiate the classifier 
clf = svm.SVC(decision_function_shape='ovo')

# fit the model
clf.fit(X_train, y_train)

# Predict the Test set results
y_pred = clf.predict(X_test)

print("The accuracy score of SVM Classifier is {}".format(accuracy_score(y_test, y_pred)))
print()
print("Classification Report:")
print(classification_report(y_test, y_pred))
print()
```
